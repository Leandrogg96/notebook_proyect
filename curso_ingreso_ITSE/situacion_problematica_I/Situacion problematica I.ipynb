## Lectura de la planilla colaborativa en formato excel y almacenamiento en dataframe

```python
# Importar libreria pandas 
import pandas as pd

# Creo el dataframe sobre la planilla colaborativa
excel_path = r"C:\Users\rig1\Desktop\Leandro\jupyter_notebook\TSCDIA-CIng-SP1.xlsx"
dataframe = pd.read_excel(excel_path, "Datos")
dataframe.style
```

## Cálculo de la media, desviación estándar y coeficiente de variabilidad.

### Edad

#### - Cálculo de la media:

```python
# Calculo de la media usando el metodo 'mean()'
media_edad = dataframe["Edad"].mean()
print(f'La media es: {media_edad:.2f}')
```

#### - Cálculo de la desviación estándar:

```python
# Calculo de la deviacion estandar usando el metodo 'std()'
dev_estandar_edad = dataframe["Edad"].std()
print(f'La deviacion estandar es: {dev_estandar_edad:.2f}')
```

#### - Cálculo del coeficiente de variabilidad:

```python
# Calculo del coeficiente de variabilidad usando los valores antes calculados
coef_varia_edad = (dev_estandar_edad / media_edad) * 100
print(f'El coficiente de variabilidad es: {coef_varia_edad:.2f}%')
```

#### Conclusión

##### Podemos obervar una variabilidad del 29.73%. La desviación estándar es cerca del 30% de la media, lo que indica que hay una cantidad razonable de dispersión en los datos. Al encontrarse dentro del rango considerado moderado (20% - 30%) podemos concluir que los datos presentan una dispersión razonable alrededor de la media. Se dice entonces que la variabilidad es moderada.

### Distancia

## Gráfico de barras

### Modos de transporte

```python
# Importo librerias que usare para la creacion del grafico de barras
import re
import unicodedata
import seaborn as sns
import matplotlib.pyplot as plt

# Funcion que me permite unificar los modos de transporte (no considerando: mayusculas, acento, espacios, putnos, comas)
def limpiar_texto(texto):
    if isinstance(texto, str):  # Asegurarse de que el texto sea una cadena
        # Convertir a minusculas
        texto = texto.lower()
        # Normalizar y eliminar acentos
        texto = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('ASCII')
        # Eliminar signos de puntuacion y espacios extras
        texto = re.sub(r'[^\w\s]', '', texto)
        texto = re.sub(r'\s+', ' ', texto).strip()
    else:
        texto = ''  # O reemplazar por un valor predeterminado
    return texto

# Aplicar la funcion de limpieza a la columna 'Modo de Transporte'
dataframe['Modo de Transporte'] = dataframe['Modo de Transporte'].apply(limpiar_texto)

# Contar la frecuencia de cada modo de transporte
modo_transporte_counts = dataframe['Modo de Transporte'].value_counts().reset_index()
modo_transporte_counts.columns = ['Modo de Transporte', 'Cantidad']

# Colores para el grafico de barras (1 color para cada modo)
colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#e377c2']

# Crear el grafico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x='Modo de Transporte', y='Cantidad', data=modo_transporte_counts, palette=colores, hue='Modo de Transporte', dodge=False)
plt.xlabel('Modo de Transporte')
plt.ylabel('Cantidad')
plt.title('Frecuencia de Modos de Transporte')
plt.xticks(rotation=45)
plt.show()
```
